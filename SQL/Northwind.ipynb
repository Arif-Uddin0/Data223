## Northwind

How many employees have a home city of London?

```sql
USE Northwind;

SELECT COUNT(*) AS From_London
FROM Employees
WHERE city = 'London';
```

Which employee is a Doctor?

```sql
USE Northwind;

SELECT CONCAT(FirstName, ' ', LastName) AS Name
FROM Employees
WHERE TitleOfCourtesy = 'Dr.';
```

How many products are discontinued?

```sql
USE Northwind;

SELECT COUNT(*)
FROM Products
WHERE Discontinued = 1;
```

What are the names and product ID's of the products with unit price below 5?

```sql
USE Northwind;

SELECT ProductID, ProductName, UnitPrice 
FROM Products
WHERE UnitPrice < 5;
```

Which categories have a category name with initials beginning with B or S?

```sql
USE Northwind;

SELECT *
FROM Categories
WHERE CategoryName LIKE 'B%' OR CategoryName LIKE 'S%'
```

How many orders are there for EmployeeID's 5 and 7 (Total)?

```sql
USE Northwind;

SELECT EmployeeID, COUNT(EmployeeID)
FROM Orders
WHERE EmployeeID IN (5,7)
GROUP BY EmployeeID;
```

Concatenate first and last name of Employees table

```sql
USE Northwind;

SELECT CONCAT(FirstName, ' ', LastName) AS Employee_Name
FROM Employees;
```

List the 6 countries that have Region codes in the Customers Table

```sql
USE Northwind;

SELECT DISTINCT Country
FROM Customers
WHERE Region IS NOT NULL;
```

Add a new column to show the Net Total which has the discount applied

```sql
USE Northwind;

SELECT UnitPrice, Quantity, Discount, (UnitPrice * Quantity)*(1-Discount) AS "Net Total"
FROM [Order Details]
```

Use ORDER BY to identify the highest Net Total in the Order Details table

```sql
USE Northwind;

SELECT (UnitPrice * Quantity)*(1-Discount) AS "Net Total"
FROM [Order Details]
ORDER BY [Net Total] DESC;
```

Two Order No. with the highest total

```sql
USE Northwind;

SELECT TOP 2 OrderID
FROM [Order Details]
ORDER BY (UnitPrice * Quantity)*(1-Discount) DESC
```

Use CHARINDEX to list only Product Names that contain a single quote

```sql
USE Northwind;

SELECT ProductName
FROM Products
WHERE CHARINDEX('''', ProductName) > 0;
```

Output a list of Employees from the Employees table including their Name and Age

```sql
USE Northwind;

SELECT CONCAT(FirstName, ' ', LastName) AS Name, DATEDIFF(YEAR, BirthDate, GETDATE())
FROM Employees
```

Use CASE to add column for Retirement Statuses

```sql
USE Northwind;

SELECT CONCAT(FirstName,' ',LastName) AS FullName, DATEDIFF(YEAR,BirthDate,GETDATE()) AS Age,
CASE WHEN DATEDIFF(YEAR,BirthDate,GETDATE()) > 60 THEN 'Retirement Due'
WHEN DATEDIFF(YEAR,BirthDate,GETDATE()) > 65 THEN 'Retired'
ELSE 'More than 5 years to go'
END AS "Retirement Status"
FROM Employees
```

GROUP BY to calculate Average Reorder Level for all products by Category ID

```sql
USE Northwind;

SELECT AVG(ReorderLevel) AS Average
FROM Products
GROUP BY CategoryID
ORDER BY Average DESC;
```

GROUP BY Supplier showing an average of Units on Order for each Supplier

```sql
USE Northwind;

SELECT Suppliers.companyname, AVG(Products.unitsonorder) AS avg_units_on_order
FROM Products
INNER JOIN Suppliers ON Products.supplierid = Suppliers.supplierid
GROUP BY Suppliers.companyname;
```

```sql
USE Northwind;

SELECT Orders.orderid, Orders.orderdate, Orders.freight, Customers.companyname, CONCAT(Employees.firstname, ' ', Employees.lastname) AS employee_name
FROM Orders
INNER JOIN Customers ON Orders.customerid = Customers.customerid
INNER JOIN Employees ON Orders.employeeid = Employees.employeeid;
```

Using Subquery in the WHERE clause

```sql
USE Northwind;

SELECT orderid, productid, unitprice, quantity, discount
FROM [Order Details]
WHERE productid IN (SELECT productid
FROM [Order Details]
WHERE discount > 0
);
```

Simple JOIN

```sql
USE Northwind;

SELECT OD.orderid, OD.productid, OD.unitprice, OD.quantity, OD.discount
FROM [Order Details] OD
JOIN (SELECT DISTINCT productid
FROM [Order Details]
WHERE discount > 0
) PD
ON OD.productid = PD.productid;
```
