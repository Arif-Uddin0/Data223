```sql
CREATE DATABASE Sparta;
```

```sql
USE Sparta;
```

```sql
DROP TABLE IF EXISTS courses, trainers, modules, trainees, clients;
```

```sql
CREATE TABLE Courses (
    course_ID INT PRIMARY KEY IDENTITY (1,1),
    course_name VARCHAR(30) NOT NULL,
    course_difficulty VARCHAR(30),
    course_type VARCHAR(30),
    trainer_id INT NOT NULL,
    trainee_id INT NOT NULL
);

CREATE TABLE Trainers (
    trainer_id INT PRIMARY KEY IDENTITY (100,10),
    trainer_name VARCHAR(30) NOT NULL,
    trainer_qualification VARCHAR(50),
    course_id INT NOT NULL,
    trainee_id INT NOT NULL
);

CREATE TABLE Modules (
    module_id INT PRIMARY KEY IDENTITY (50,2),
    module_name VARCHAR(50) NOT NULL,
    module_time VARCHAR(30) NOT NULL,
    course_id INT NOT NULL
); 

CREATE TABLE Trainees(
trainee_id INT PRIMARY KEY IDENTITY(1,1),
trainee_name VARCHAR(50) NOT NULL,
phone_number VARCHAR(20) NOT NULL,
email_address TEXT,
city VARCHAR(20),
course_id INT
);

CREATE TABLE Clients(
client_id INT PRIMARY KEY IDENTITY(1,1),
client_name VARCHAR(50) NOT NULL,
date_joined DATE,
trainee_id INT
); 
```
