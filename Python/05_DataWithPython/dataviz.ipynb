```python
df_cal = pd.read_csv('ams_calendar.csv')
df_cal.columns
```

```python
df_cal.head()
```

```python
# What are the busiest times of the year to visit?

df_cal['date'] = pd.to_datetime(df_cal['date'])
df_cal['month'] = df_cal['date'].dt.month
df_cal['year'] = df_cal['date'].dt.year

monthly_counts = df_cal[df_cal['available'] == 'f'].groupby('month')['available'].count().reset_index()

sns.barplot(x='month', y='available', palette='autumn', data=monthly_counts)

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(0, 12), month_names)

plt.title('Busiest Times of the Year to Visit')
plt.xlabel('Month')
plt.ylabel('Number of Listings Booked')

plt.show()
```

```python
# By how much do prices spike?

monthly_price_spikes = df_cal.groupby('month')['price'].agg('mean').reset_index()

sns.barplot(x='month', y='price', palette='winter_r', data=monthly_price_spikes)

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(0, 12), month_names)

plt.title('Price Spikes by Month')
plt.xlabel('Month')
plt.ylabel('Price Spike')

plt.show()
```

```python
# By how much do prices spike?

df_cal['price'] = pd.to_numeric(df_cal['price'].str.replace('$', '').str.replace(',', ''))

monthly_price_spikes = df_cal.groupby('month')['price'].apply(lambda x: x.max() - x.min()).reset_index()

sns.barplot(x='month', y='price', palette='winter_r', data=monthly_price_spikes)

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(0, 12), month_names)

plt.title('Price Spikes by Month')
plt.xlabel('Month')
plt.ylabel('Price Spike')

plt.show()
```

```python
sns.boxplot(x='month', y='price', data=df_cal)

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(0, 12), month_names)

plt.title('Price Spikes by Month')
plt.xlabel('Month')
plt.ylabel('Price')

plt.show()
```

```python
display(df_cal.price.min())
display(df_cal.price.max())
```

```python
# Average Price by Availability

availability_prices = df_cal.groupby('available')['price'].mean().reset_index()

colours = ['darkgreen', 'darkred']
sns.barplot(x='available', y='price', palette=colours, order=('t','f') , data=availability_prices)

available = ['True', 'False']
plt.xticks(range(0, 2), available)

plt.title('Average Price by Availability')
plt.xlabel('Availability')
plt.ylabel('Average Price')

plt.show()
```
