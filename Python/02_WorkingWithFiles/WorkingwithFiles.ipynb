Flat Files
Flat files are basic text files containing records (rows-row of fields or attributes) of data without structured relationships between the records (no joins).
Flat files are the most common way of storing data for import and export from databases.
Flat files can be comma separated (CSV) or tab separated (TSV).
Flat files can be read line by line or all at once.
Flat files can be read using the csv module, NumPy module or the Pandas module.
Flat files can be read using the open() function.

Return list of first name, last name and email

```python
import csv

def print_user_data(filename):
    with open(filename, 'r') as usr_data:
        reader = csv.reader(usr_data)
        next(reader)
        for row in reader:
            first_name, last_name, email = row[0], row[1], row[2]
            print([first_name, last_name, email])

print_user_data('new_user_data.csv')
```

# Generator

```python
import csv

def get_user_data(path):
    with open(path, 'r') as usr_data:
        reader = csv.reader(usr_data)
        next(reader)
        for row in reader:
            first_name, last_name, email = row[0], row[1], row[2]
            yield first_name, last_name, email

user_data_gen = get_user_data('new_user_data.csv')
for first_name, last_name, email in user_data_gen:
    print([(first_name), (last_name), (email)])
```

# IMDB data transformation

The task is to transform the the IMDB data using the below criteria:

only the films from 2010 onwards should be selected
movies should be the only title type
only the primary title should be used
remove the end date and isadult fields entirely

```python
import csv

def csv_file():

    with open('imdbtitles.csv','r') as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0] == 'movie' and int(row[4]) >= 2010:
                print(row[0],row[1],row[4],row[6],row[7])

csv_file()
```

```python
import pandas as pd

imdb = pd.read_csv('imdbtitles.csv')

movies = imdb[(imdb['titleType'] == 'movie') & (imdb['startYear'] >= 2010)][['primaryTitle', 'runtimeMinutes', 'genres', 'startYear']]

movies = movies.drop(['startYear', 'runtimeMinutes'], axis='columns')

print(movies)
```
