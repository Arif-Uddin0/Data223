```python
from google.cloud import storage
import os 
```

```python
# create a credentials object using the key file
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]='GCP.json'

bucket_name = 'arif-sparta223' # Name of the bucket in GCP

# create a client object
storage_client = storage.Client()

# list files in bucket
for bucket in storage_client.list_buckets(): # list all buckets in GCP 
    print("Bucket name:",bucket.name)
    for file in bucket.list_blobs():
        print("\tFile name:",file.name)
        print("\tFile size:",file.size,"bytes")
```

```python
# write a function 
def download_blob(bucket_name):
    """Downloads a blob from the bucket."""
    storage_client = storage.Client()  # create a client object 
    bucket = storage_client.get_bucket(bucket_name)  # get bucket object 

    # prompt user to input source blob name and destination file name
    source_blob_name = input("Enter source blob name: ")
    destination_file_name = source_blob_name

    blob = bucket.blob(source_blob_name)  # create a new Blob object 
    blob.download_to_filename(destination_file_name)  # download file to a destination 
    print('Blob {} downloaded to {}.'.format(source_blob_name, destination_file_name))  
    return destination_file_name

download_blob(bucket_name)
```

```python
# return the content of the blob and download as a string
def download_blob_content(bucket_name, source_blob_name):
    """Downloads a blob from the bucket and returns its content as a string."""
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)
    blob = bucket.blob(source_blob_name)

    content = blob.download_as_string()
    return content
 
content = download_blob_content(bucket_name, input('Enter blob name in full: '))
print(content)
```

```python
def upload_file_to_bucket(bucket_name, source_file_name, destination_blob_name):
    """Uploads a file from the local directory to the bucket."""
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)

    blob.upload_from_filename(source_file_name)
    print(f'File {source_file_name} uploaded to {destination_blob_name} in {bucket_name} bucket.')

upload_file_to_bucket(bucket_name, input('Enter blob name in full: '), input('Enter destination blob name in full: '))
```
